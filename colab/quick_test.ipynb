```python
# Check output files
!ls -lh output/

# Show channel count
import os
for fname in os.listdir('output'):
    if fname.endswith('.m3u'):
        path = os.path.join('output', fname)
        with open(path) as f:
            channels = sum(1 for line in f if line.startswith('#EXTINF'))
        size = os.path.getsize(path) / (1024 * 1024)
        print(f"\n📄 {fname}")
        print(f"   Channels: {channels:,}")
        print(f"   Size: {size:.2f} MB")
```

```python
# Preview first 20 channels
print("=== First 20 channels from combined playlist ===\n")
with open('output/combined_by_country.m3u') as f:
    lines = f.readlines()
    
channel_count = 0
for line in lines:
    line = line.strip()
    if line.startswith('#EXTINF'):
        channel_count += 1
        # Extract channel name
        if ',' in line:
            name = line.split(',')[-1]
        print(f"#{channel_count}: {name}")
    elif line and not line.startswith('#') and channel_count > 0:
        # Show URL (truncated)
        url_display = line[:80] + '...' if len(line) > 80 else line
        print(f"     {url_display}")
        print()
        
    if channel_count >= 20:
        break
```

```python
# Download the combined playlist
from google.colab import files

print("📥 Downloading combined playlists...")

if os.path.exists('output/combined_by_country.m3u'):
    files.download('output/combined_by_country.m3u')

if os.path.exists('output/checked_combined_by_country.m3u'):
    files.download('output/checked_combined_by_country.m3u')

print("\n✅ Done! Open the downloaded .m3u file in your IPTV player.")
```
